```python
#Read the data with the smiles from the github
import pandas as pd
data_calc=pd.read_csv('data_smi.csv',low_memory=False)
```

```python
#Visualize the first rows of the dat
data_calc.head()
```

```python
from rdkit import Chem
from mordred import Calculator, descriptors
```

```python
#define the variable to calculate the descriptors, 3D descriptors are ignore
calc=Calculator(descriptors,ignore_3D=True)
```

```python
#Checking how many descriptors we have
print(f'There will be calculated {len(calc.descriptors)} descriptors')
```

```python
import pandas as pd
from rdkit.Chem import PandasTools
```

```python
#Reading the smiles from the columnn 'SMILES' and converting to mol files
mols = [Chem.MolFromSmiles(s) for s in data_calc['SMILES']]
```

```python
#Show how many molecules were loaded
print(f'Number of molecules: {len(mols)}')
```

```python
#Proceed with the calculation
df = calc.pandas(mols)
```

```python
#To visualize our table with the calculations
df.head()
```

```python
#Concatenate the name from smiles csv to the output calculated descriptors
data_concat =pd.concat([data_calc.iloc[:,0],df],axis=1)
```

```python
#Visualize the data set
data_concat.head()
```
